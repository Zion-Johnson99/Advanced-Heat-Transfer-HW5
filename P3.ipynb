```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_network():
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.set_xlim(0, 14)
    ax.set_ylim(0, 8)
    ax.axis('off')

    # 节点样式 - 使用bbox参数来设置文本边框
    node_props = dict(bbox=dict(boxstyle="circle,pad=0.3", facecolor="white", edgecolor="black"), 
                      ha='center', va='center')
    
    # 绘制节点
    ax.text(1, 4, "$E_{b1}$", **node_props)
    ax.text(3, 4, "$J_1$", **node_props)
    ax.text(7, 4, "$J_{2i}$", **node_props)
    ax.text(9, 4, "$E_{b2}$", **node_props)
    ax.text(5, 7, "$E_{be}$\n(Openings)", **node_props) # 上方环境节点
    ax.text(11, 4, "$E_{be}$\n(Outer Surr)", **node_props) # 右侧环境节点

    # 绘制连线和电阻
    # 1. Eb1 -> J1
    ax.annotate("", xy=(2.7, 4), xytext=(1.3, 4), arrowprops=dict(arrowstyle="-"))
    ax.text(2, 3.5, r"$\frac{1-\epsilon_1}{\epsilon_1 A_1}$", ha='center')

    # 2. J1 -> J2i (Main exchange)
    ax.annotate("", xy=(6.7, 4), xytext=(3.3, 4), arrowprops=dict(arrowstyle="-"))
    ax.text(5, 3.5, r"$\frac{1}{A_1 F_{12}}$", ha='center')

    # 3. J1 -> Ebe (To Openings)
    ax.annotate("", xy=(4.8, 6.7), xytext=(3.2, 4.3), arrowprops=dict(arrowstyle="-"))
    ax.text(3.5, 5.5, r"$\frac{1}{A_1 F_{1e}}$", ha='center')

    # 4. J2i -> Ebe (To Openings)
    ax.annotate("", xy=(5.2, 6.7), xytext=(6.8, 4.3), arrowprops=dict(arrowstyle="-"))
    ax.text(6.5, 5.5, r"$\frac{1}{A_2 F_{2e}}$", ha='center')

    # 5. J2i -> Eb2 (Inner surface resistance)
    ax.annotate("", xy=(8.7, 4), xytext=(7.3, 4), arrowprops=dict(arrowstyle="-"))
    ax.text(8, 3.5, r"$\frac{1-\epsilon_2}{\epsilon_2 A_2}$", ha='center')
    
    # 6. Eb2 -> Ebe_outer (Outer Radiation)
    # 这里通常简化为直接辐射，因为外表面对着大空间
    ax.annotate("", xy=(10.7, 4), xytext=(9.3, 4), arrowprops=dict(arrowstyle="-"))
    ax.text(10, 3.5, r"$\frac{1}{\epsilon_0 A_{2}}$", ha='center') # 包含了表面热阻和空间热阻(F=1)

    plt.title("Radiation Network for Concentric Cylinders with Open Ends")
    plt.show()

if __name__ == "__main__":
    draw_network()
```
